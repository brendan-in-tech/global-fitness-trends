```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#inputs
workout_df = pd.read_csv('data/workout.csv')
keywords_df = pd.read_csv('data/three_keywords.csv')
geo_df = pd.read_csv('data/workout_geo.csv')
keywords_geo_df = pd.read_csv('data/three_keywords_geo.csv')
```

```python
def normalize(df):
    """normalizes based on
        - year
        - sums metrics
    """
    df = df.dropna()
    df['year'] = df["month"].apply(lambda x: x.split('-')[0])
    df = df.groupby('year').sum()
    return df
```

```python
workout_df_normalized = normalize(workout_df)
workout_df_normalized.plot()
plt.show()
```
